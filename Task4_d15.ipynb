###### Department wise toppers (horizantal bar graph or any visual representations of your choice)

```python
#college 1
cumulative_score = college_1_df['python'] + college_1_df['mysql'] + college_1_df['Previous Geekions'] + college_1_df['CodeKata Score'] + college_1_df['Rising']
college_1_df['cumulative_score'] = cumulative_score

college_1_df.groupby('Department').max()
```

```python
df1 = college_1_df.groupby('Department').max()
df1 = df1.reset_index()
Topper = df1[['Name','Department']].agg('_'.join, axis = 1)
df1['Topper'] = Topper
df1
```

```python
colors = ['blue','green','brown']
ax = df1.plot.barh('Topper','cumulative_score',color = colors, legend = False)
ax.set_xlabel('cumulative_score')
for p in ax.patches:
    ax.annotate("%.2f" % p.get_width(), (p.get_x() + p.get_width(), p.get_y()), xytext=(5, 10),textcoords='offset points')
```

```python
# college 2
cumulative_score = college_2_df['python'] + college_2_df['mysql'] + college_2_df['Previous Geekions'] + college_2_df['CodeKata Score'] + college_2_df['Rising'] + college_2_df['python_en'] + college_2_df['mysql'] + college_2_df['computational_thinking']
college_2_df['cumulative_score'] = cumulative_score

df2 = college_2_df.groupby('Department').max()
df2
```
